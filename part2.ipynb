```python
"""
Plot the percentage of the instances in each output class and store the graphic in a file called penguin-classes.gif
/ abalone-classes.gif.
"""
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
# Load Penguin dataset
penguin_path = r'C:\Users\ibrah\OneDrive\Desktop\COMP472-A1-datasets\penguins.csv'
penguin_data = pd.read_csv(penguin_path)
```

```python
# Plot and save the percentage of instances in each output class for Penguin dataset
penguin_class_percentages = penguin_data['species'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 6))
penguin_class_percentages.plot(kind='bar', color=['blue', 'green', 'orange'])
plt.title('Penguin Class Percentages')
plt.xlabel('Species')
plt.ylabel('Percentage')
plt.savefig('penguin-classes.png')
plt.show()
```

```python
# Load Abalone dataset
abalone_path = r'C:\Users\ibrah\OneDrive\Desktop\COMP472-A1-datasets\abalone.csv'
abalone_data = pd.read_csv(abalone_path)
```

```python
# Plot and save the percentage of instances in each output class for Abalone dataset
abalone_class_percentages = abalone_data['Type'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 6))
abalone_class_percentages.plot(kind='bar', color=['blue', 'green', 'orange'])
plt.title('Abalone Class Percentages')
plt.xlabel('Type')
plt.ylabel('Percentage')
plt.savefig('abalone-classes.png')
plt.show()
```
